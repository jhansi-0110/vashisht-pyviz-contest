# Mammals Schedule 1


## 1. Conservation Status Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Load the Excel file
file_path = "AIWC-DATA-COLLECT.xlsx"
sheet_name = "Mammals Schedule-I"

# Read the Excel sheet into a DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 1. Conservation Status Analysis
plt.figure(figsize=(12, 6))
status_counts = df["body/conservationStatus"].value_counts()
status_counts.plot(kind='bar', color=['#e63946', '#f4a261', '#2a9d8f', '#457b9d', '#1d3557'])
plt.title("Species Distribution by Conservation Status", fontweight='bold')
plt.xlabel("Conservation Status", fontweight='bold')
plt.ylabel("Number of Species", fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## 2. Taxonomic Distribution (Class level)

```python
# 2. Taxonomic Distribution (Class level)
plt.figure(figsize=(10, 6))
taxonomy_counts = df["body/taxonomy/class"].value_counts()
taxonomy_counts.plot(kind='pie', autopct='%1.1f%%', 
                    colors=['#ffbe0b', '#fb5607', '#ff006e', '#8338ec', '#3a86ff'],
                    startangle=90)
plt.title("Species Distribution by Class", fontweight='bold')
plt.ylabel("")
plt.tight_layout()
plt.show()
```

## 3. Habitat Distribution Analysis

```python
# 3. Habitat Distribution Analysis
# First we need to extract all habitat values (since they're in multiple columns)
habitat_columns = [col for col in df.columns if 'geoInformation/habitats' in col]
habitats = df[habitat_columns].stack().value_counts()

plt.figure(figsize=(12, 6))
habitats.plot(kind='barh', color='#588157')
plt.title("Species Habitat Distribution", fontweight='bold')
plt.xlabel("Number of Species", fontweight='bold')
plt.ylabel("Habitat Type", fontweight='bold')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## pie chart analysis of Habitat Distribution Analysis(Alternative chart)

```python
# Count species per habitat
habitat_counts = df["geoInformation/habitats/0"].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
habitat_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, cmap='Set3')
plt.title("Habitat Distribution")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

# 4. Schedule Part Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "AIWC-DATA-COLLECT.xlsx"
sheet_name = "Mammals Schedule-I"

# Read the Excel sheet into a DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Count species per schedule part
schedule_counts = df["body/SchedulePart"].value_counts()

# Dynamically set explode values (small explosion for all slices)
explode_values = [0.05] * len(schedule_counts)

# Plot pie chart
plt.figure(figsize=(8, 6))
schedule_counts.plot(kind='pie', autopct='%1.1f%%',
                     colors=['#f72585', '#7209b7', '#3a0ca3', '#4361ee', '#4cc9f0'],
                     explode=explode_values)

plt.title("Species Distribution by Schedule Part", fontweight='bold')
plt.ylabel("")  # Hide y-axis label
plt.tight_layout()
plt.show()
```

# 6. Geographic Distribution (Top Locations)

## 3.Scatter Plot: Body Size vs. Tail Length

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the Excel file
file_path = "AIWC-DATA-COLLECT.xlsx"  # Ensure the file path is correct
sheet_name = "Mammals Schedule-II"

# Read Excel sheet
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Extract body size (handles both "30.3 cm" and "25-30 cm" formats)
df["Body Size (cm)"] = pd.to_numeric(df["descriptionOrExplanation"].str.extract(r'(\d+\.?\d*)\s*-?\s*\d*\s*cm')[0], errors='coerce')

# Extract tail length (handles "Tail size length is 15-23 cm" and "Tail of 12 cm" formats)
df["Tail Length (cm)"] = pd.to_numeric(df["descriptionOrExplanation"].str.extract(r'Tail.*?(\d+\.?\d*)\s*-?\s*\d*\s*cm')[0], errors='coerce')

# Drop rows with missing values
df_filtered = df.dropna(subset=["Body Size (cm)", "Tail Length (cm)"])

# Ensure df_filtered is not empty
if df_filtered.empty:
    print("No valid numerical data found for Body Size or Tail Length. Check the data extraction.")
else:
    # Scatter plot with axis limits adjusted
    plt.figure(figsize=(10, 6))
    plt.scatter(df_filtered["Body Size (cm)"], df_filtered["Tail Length (cm)"], alpha=0.7, color='purple')

    # Set limits based on actual data
    plt.xlim(df_filtered["Body Size (cm)"].min() - 5, df_filtered["Body Size (cm)"].max() + 5)
    plt.ylim(df_filtered["Tail Length (cm)"].min() - 5, df_filtered["Tail Length (cm)"].max() + 5)

    # Labels and title
    plt.xlabel("Body Size (cm)")
    plt.ylabel("Tail Length (cm)")
    plt.title("Body Size vs. Tail Length")
    plt.grid(True)

    plt.show()
```

## 4.Stacked Bar Chart: Taxonomic Classification

```python
# Count species by taxonomy class
taxonomy_counts = df["body/taxonomy/class"].value_counts()

# Plot bar chart
plt.figure(figsize=(12, 6))
taxonomy_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Taxonomic Class")
plt.ylabel("Number of Species")
plt.title("Species Count by Taxonomic Class")
plt.xticks(rotation=45)
plt.show()
```

# Fish sch 2

```python
# Load the Excel file
file_path = "AIWC-DATA-COLLECT.xlsx"
sheet_name = "Fish schedule -II"

df = pd.read_excel(file_path, sheet_name=sheet_name)

# Count species per conservation status
status_counts = df["body/conservationStatus"].value_counts()

# Plot conservation status distribution
plt.figure(figsize=(10, 6))
status_counts.plot(kind='bar', color=['red', 'orange', 'yellow', 'green', 'blue'])
plt.xlabel("Conservation Status")
plt.ylabel("Number of Species")
plt.title("Species Count by Conservation Status")
plt.xticks(rotation=45)
plt.show()

# Distribution of species by taxonomy class
taxonomy_counts = df["body/taxonomy/class"].value_counts()
plt.figure(figsize=(12, 6))
taxonomy_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Taxonomy Class")
plt.ylabel("Number of Species")
plt.title("Species Count by Taxonomy Class")
plt.xticks(rotation=45)
plt.show()

# Analyzing body size distribution (Filtering valid numeric values)
df_filtered = df.dropna(subset=["technicals/parts/0/typeDetails/0/value"])
df_filtered = df_filtered[df_filtered["technicals/parts/0/typeDetails/0/value"].astype(str).str.contains("cm")]
df_filtered["Size_cm"] = df_filtered["technicals/parts/0/typeDetails/0/value"].str.replace(" cm", "").astype(float)

# Plot histogram for body size
df_filtered["Size_cm"].hist(bins=10, color='purple', edgecolor='black')
plt.xlabel("Body Size (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Species by Body Size")
plt.show()

# Habitat distribution
df_melted = df.melt(value_vars=["geoInformation/habitats/0", "geoInformation/habitats/1", "geoInformation/habitats/2"],
                     var_name="Habitat Type", value_name="Habitat")

habitat_counts = df_melted["Habitat"].value_counts()
plt.figure(figsize=(12, 6))
habitat_counts.plot(kind='bar', color='teal')
plt.xlabel("Habitat")
plt.ylabel("Number of Species")
plt.title("Distribution of Species by Habitat")
plt.xticks(rotation=45)
plt.show()

# Species count by order
order_counts = df["body/taxonomy/order"].value_counts().nlargest(10)  # Top 10 orders
plt.figure(figsize=(12, 6))
order_counts.plot(kind='bar', color='coral')
plt.xlabel("Taxonomic Order")
plt.ylabel("Number of Species")
plt.title("Top 10 Species Count by Taxonomic Order")
plt.xticks(rotation=45)
plt.show()
```
